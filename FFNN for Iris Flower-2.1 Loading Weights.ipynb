# **Neural Network Architecture**

```python
from keras.models import Sequential
from keras.layers import Dense

model= Sequential() # empty sequential type neural network
model.add(Dense(64,input_dim=4,activation="relu")) # 1st Hidden layer , Dense is fully connected layer, imput_shape=... (2D-CNN) ,64= neurons, Input shape=(4,)
model.add(Dense(32,activation="relu")) # 2st Hidden layer
model.add(Dense(3,activation="softmax")) # output layer

model.compile (loss="categorical_crossentropy", optimizer="Adamax",metrics =["accuracy"])
model.summary()
```

# **Load the weight**

```python
model.load_weights('FFNN-iris.h5')
```

# **Given a prediction**

```python
import numpy as np

test_data_1=np.array([5.3,2.3,4.7,2.1]).reshape(1,-1) # convert row vecter using reshape because predict function takes 2D array
# No need reshape for multiple prediction
predicted_target_1=model.predict(test_data_1)
print(predicted_target_1)
```
