# **Load the model**

```python
from keras.models import load_model
model=load_model('FFNN-iris.model')
```

# **Given a prediction**

```python
import numpy as np

test_data_1=np.array([5.3,2.3,4.7,2.1]).reshape(1,-1) # convert row vecter using reshape

predicted_target_1=model.predict(test_data_1)
print(predicted_target_1)
```

```python
label=np.argmax(predicted_target_1)
print(label)
```
