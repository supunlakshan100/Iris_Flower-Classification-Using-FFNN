```python
import pandas as pd
import tensorflow as tf

dataset=pd.read_csv ('H:\My Drive\ANN\FFNN for Iris Flower\iris.csv').values
print(dataset,dataset.shape)
```

```python
data=dataset[:,0:4]
target=dataset[:,4]

print(data.shape,target.shape)
```

```python
import tensorflow as tf 
New_target=tf.keras.utils.to_categorical(target) #one hot encoding
New_target=tf.keras.utils.to_categorical(target) #one hot encoding
print(New_target[:10])
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data,train_target,test_target=train_test_split(data, New_target,test_size=0.1)
```

```python
print(train_data,train_data.shape)
```

```python
print(test_data,test_data.shape)
```

### Build A Neural Network

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from keras.models import Sequential
from keras.layers import Dense

model= Sequential() # empty sequential type neural network ,object, sequential is a class. There are 2 types of NN, sequential and model type
model.add(Dense(64,input_dim=4,activation="relu")) # 1st Hidden layer , Dense is fully connected layer, imput_shape=... (2D-CNN) ,64= neurons, Input shape=(4,)
model.add(Dense(32,activation="relu")) # 2st Hidden layer
model.add(Dense(3,activation="softmax")) # output layer

model.compile (loss="categorical_crossentropy", optimizer="Adamax",metrics =["accuracy"]) 
# Adamax is a type of optimizer, loss function is categorical_crossentropy, metrics is accuracy (othervise it will show only loss)
model.summary() # to see the summary of the model
```

# **get_weights**

```python
from sklearn.utils.validation import Parameter
for layer in model.layers:
  print("============",layer.name,"=============")

  parameters=layer.get_weights()
  print(parameters)
```

###  Trainning the Model

```python
model.fit(train_data,train_target, epochs=1500, validation_split=0.1) 
# validation_split=0.1 means 10% of the training data will be used for validation, unseen data( like tearm test marks)
# looking real time accuracy using validation_split=0.1, times for each epoch=1500
```

```python
from matplotlib import pyplot as plt

plt.plot(model.history.history['loss'],label= 'loss')
plt.plot(model.history.history['val_loss'],label='val_loss')
plt.legend()
plt.xlabel('#epoch')
plt.ylabel('loss')
plt.show
```

```python
plt.plot(model.history.history['accuracy'],label='accuracy')
plt.plot(model.history.history['val_accuracy'],label='val_accuracy')
plt.legend()
plt.xlabel('#epoch')
plt.ylabel('loss')
plt.show
```
